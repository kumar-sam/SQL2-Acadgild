```python
# Assignment 13.1
"""
Read the following data set:
https://archive.ics.uci.edu/ml/machine-learning-databases/adult/adult.data
Task:
1. Create an sqlalchemy engine using a sample from the data set
"""
from sqlalchemy import create_engine #importing create_engine

# create a engine that store data in the memory
engine = create_engine('sqlite:///:memory:', echo=False)

from sqlalchemy.ext.declarative import declarative_base
Base = declarative_base() # constructive a base class

from sqlalchemy import Column, Integer, String

class Adult(Base): #class definition which sql-chemy uses to create the table
    __tablename__ = 'adult' # tablename adult
    
    id=Column(Integer, primary_key=True) #int id
    age =Column(Integer) #int age
    workclass=Column(String) #varchar workclass
    fnlwgt=Column(Integer) #integer fnlwgt
    education=Column(String) #varchar education
    education_num=Column(Integer) #integer eduction
    marital_status=Column(String) #varchar martial_status
    occupation=Column(String) #varchar occupation
    relationship=Column(String) #varchar relationship
    race=Column(String) #varchar race
    sex=Column(String) #varchar sex
    capital_gain=Column(Integer) #integer capital_gain
    capital_loss=Column(Integer) #integer capital_loss
    hours_per_week=Column(Integer) #integer hours_per_week
    native_country=Column(String) #varchar native_country
    classs=Column(String) #varchar classs
    
    def __repr__(self):
        return "<Adult(id='%d')>" % (self.id)
    
# Create all tables in the engine. This is equivalent to "Create Table" statements in raw SQL.    
Base.metadata.create_all(engine)
```

```python
# Assignment 13.2
"""
2. Write two basic update queries
"""
# update-1
from sqlalchemy import update   #importing the library
update(Adult).where(Adult.education=='Assoc-voc').values(native_country='Senior') # update table where education='Assoc-voc'
session.commit()


# update-2
update(Adult).where(Adult.race=='Amer-Indian-Eskimo').values(classs ='1111K') # update table where race=='Amer-Indian-Eskimo'
session.commit()
```

```python
# Assignment 13.3
"""
2. Write two delete queries
"""
# deletion-1
session.query(Adult).filter(Adult.id == 10).delete() #deleting row where adult.id=10
session.commit()  #commiting the session
print(session.query(Adult).all())  #printing all the outputs of query

# deletion-2
#deleting tuple where race='Asian-Pac-Islander' and occupation=='Craft-repair'
session.query(Adult).filter(Adult.race=='Asian-Pac-Islander').filter(Adult.occupation=='Craft-repair').delete()
session.commit()  # commiting the session
print(session.query(Adult).all())  #printing all the outputs of the query
```

```python
# Assignment 13.4

"""filter query-1"""

# printing tuples where native_country=='United-States' and sex=='Male'
for adult in session.query(Adult).filter(Adult.native_country=='United-States').filter(Adult.sex=='Male'):
    print(adult)
    
"""filter query-2"""
# printing tuples where education like('%helors') order by Adult.id
query = session.query(Adult).filter(Adult.education.like('%helors')).order_by(Adult.id)
for item in query.all():
    print(item)    
```
